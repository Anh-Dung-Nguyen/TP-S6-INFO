## Algorithme de Kruskal


Rappel sur l'algorithme de Kruskal :
- Il calcule un un arbre couvrant de poids minimal
- Procédure
    - Prendre un graphe non orienté pondéré
    - Sélectionner les arêtes de poids minimal sans créer de cycle
    - À la fin : un arbre couvrant (minimum spanning tree, MST)

<font color=orange>**Question** : *Proposer une implémentation de l'algorithme de Kruskal.*</font>

Rappel des étapes générales de l’algorithme
1) Trier toutes les arêtes du graphe par poids croissant
2) Initialiser chaque sommet dans un ensemble distinct (chacun est son propre parent)
3) Parcourir les arêtes triées :
- Si les deux sommets d’une arête ne sont pas dans le même ensemble → ajouter l’arête au MST
- Sinon → ignorer (elle créerait un cycle)
4) Continuer jusqu’à ce que l’arbre ait n - 1 arêtes

````python
# La fonction renvoie
# mst : la liste des arêtes retenues dans l’arbre couvrant
# total_cost : la somme des poids des arêtes du MST
def kruskal(graph_edges):
    mst = [] # Initialisation de la liste des arêtes dans l'arbre H=0
    total_cost = 0 # Initialisation du coût 
    # Compléter
    
    vertices = set()
    for u, v, _ in graph_edges:
        vertices.add(u)
        vertices.add(v)
        
    # chacun est son propre parent
    parent = {v: v for v in vertices}
    rank = {v: 0 for v in vertices}
    
    # trier tous les arêtes par poids croissant
    sorted_graphe = sorted(graph_edges, key = lambda graphe: graphe[2])

    # parcourir les arêtes triées
    for u, v, weight in sorted_graphe:
        x, y = u, v
        
        # chacun est son propre parent
        while parent[x] != x:
            x = parent[x]
        while parent[y] != y:
            y = parent[y]
        
        # Si les deux sommets d’une arête ne sont pas dans le même ensemble
        if x != y:
            mst.append((u, v, weight))
            total_cost = total_cost + weight
            
            if rank[x] > rank[y]:
                parent[y] = x
            elif rank[x] < rank[y]:
                parent[x] = y
            else:
                parent[y] = x
                rank[x] = rank[x] + 1
        
        # arrête jusqu’à ce que l’arbre ait n - 1 arêtes
        if len(mst) == len(vertices) - 1:
            break
    
    return mst, total_cost


##############################
# Test
edges_Test = [
    ('A', 'B', 4),
    ('A', 'C', 1),
    ('B', 'C', 3),
    ('B', 'D', 2),
    ('C', 'D', 5)
]
mst, cost = kruskal(edges_Test)
print("Arêtes de l'arbre couvrant :", mst)
print("Coût total :", cost)

# Arêtes de l'arbre couvrant : [('A', 'C', 1), ('B', 'D', 2), ('B', 'C', 3)]
# Coût total : 6
````

```
 graph mon_graphe {
     a -- b -- c;
     b -- d;
 }
 ```

 Exemple de graphe orienté au format dot :
```
 digraph mon_graphe {
     a -> b -> c;
     b -> d;
 }
 ```
